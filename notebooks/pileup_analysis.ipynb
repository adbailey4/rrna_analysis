```python
def parse_read_bases_pileup(pileup_str):
    """Parse pileup str for useful count data  
    file_specs from: http://www.htslib.org/doc/samtools-mpileup.html
    """
    matches = pileup_str.count(".")
    inserts = re.findall("\+[0-9]+[ACGTNacgtn]+", pileup_str)
    deletes = re.findall("\-[0-9]+[ACGTNacgtn]+", pileup_str)
    more_deletes = pileup_str.count("*")
    more_inserts = pileup_str.count(">")
    mismatches = re.findall("[.]([ACGTNacgtn])", pileup_str)
    my_total = np.sum([matches, len(inserts), len(deletes), len(mismatches), more_deletes, more_inserts])
    print(matches, len(inserts), len(deletes), len(mismatches), more_deletes, more_inserts, my_total, coverage)
```
